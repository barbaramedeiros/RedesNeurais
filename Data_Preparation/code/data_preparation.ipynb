```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(
    r'C:\Users\Barbara\Desktop\Estudos\Infnet\Data_Science\7-Redes_Neurais\Projeto\Data_Preparation\data\raw\winequalityN.csv')
data.head()
```

```python
data.describe()
```

```python
data.columns = data.columns.str.replace(" ", "_")
data.head()
```

```python
data.hist(figsize=(14, 12), bins=20)
plt.show()
```

```python
data['opinion'] = data['quality'].apply(lambda x: 1 if x >= 6 else 0)
data['opinion'].value_counts(normalize=True)
data = data.drop(['quality'], axis=1)
data
```

```python
dt = data['type']
dt = dt.apply(lambda x: 1 if x == 'white' else 0)
data['type'] = dt
data
```

```python
dupe = data[data.duplicated() == True]
dupe
data.drop_duplicates()
```

```python
for col in data.columns:
  if data[col].isnull().sum() > 0:
    data[col] = data[col].fillna(data[col].mean())

data.isnull().sum()
```

```python
drop_data = data[(data['total_sulfur_dioxide'] > 250)
                 & (data['residual_sugar'] >= 5)]

drop_data = data[(data['total_sulfur_dioxide'] > 200)
                 & (data['residual_sugar'] < 5)]

to_drop = drop_data.index

for t in to_drop:
    data.drop(index=t, axis=1, inplace=True)
data
```

```python
branco = data[data['type'] == 1]
branco
```

```python
t_branco = branco.copy(deep=True)

t_branco['residual_sugar'] = np.sqrt(t_branco['residual_sugar'])
t_branco['chlorides'] = np.sqrt(t_branco['chlorides'])
t_branco['free_sulfur_dioxide'] = np.log(t_branco['free_sulfur_dioxide'])
t_branco['total_sulfur_dioxide'] = np.log(t_branco['total_sulfur_dioxide'])

t_branco.hist(figsize=(14, 12), bins=30)
plt.show()
```

```python
t_branco.to_csv(
    r"C:\Users\Barbara\Desktop\Estudos\Infnet\Data_Science\7-Redes_Neurais\Projeto\Data_Preparation\data\processed\branco.csv", index=False)
```
